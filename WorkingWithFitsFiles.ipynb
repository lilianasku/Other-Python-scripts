### Work with Binary Fits Files

How to work with FITS files used in many large-scale scientific data projects.

The Flexible Image Transport System (FITS) is an open-standard digital file format used for storage, transmission and processing of scientific data. It is mainly used in astronomy for processing images and tables. 

It is designed to have readable ASCII headers, describing the data, followed by binary data blocks. The headers play a role of a Metadata, while the binary sections retain high-precision of the raw scientific values. 

Fits files for this Notebook originate from the Gaia astronomical survey: https://gea.esac.esa.int/archive/

```python
#Import libraries
from astropy.io import fits
```

```python
#Read the Header of a fits file:

hdul = fits.open('./Gaia/1539145781741O-result.fits')
```

```python
#Print the structure of the FITS file to understand its contents:
hdul.info()
```
